When a share of common stock of some company is sold, the capital gain (or, sometimes, loss) is the difference between the share’s selling price and the price originally paid to buy it. This rule is easy to under- stand for a single share, but if we sell multiple shares of stock bought over a long period of time, then we must identify the shares actually being sold. A standard accounting principle for identifying which shares of a stock were sold in such a case is to use a FIFO protocol—the shares sold are the ones that have been held the longest (indeed, this is the default method built into several personal finance software packages). For example, suppose we buy $100$ shares at $\$20$ each on day 1, $20$ shares at $\$24$ on day 2, $200$ shares at $\$36$ on day 3, and then sell $150$ shares on day 4 at $\$30$ each. Then applying the FIFO protocol means that of the $150$ shares sold, $100$ were bought on day 1, $20$ were bought on day 2, and $30$ were bought on day 3. The capital gain in this case would therefore be $100·10 + 20·6 + 30·(−6)$, or $\$940.$

````python
# YOUR CODE HERE
#from ..exceptions import Empty
import re

class ArrayQueue:
  """FIFO queue implementation using a Python list as underlying storage."""
  DEFAULT_CAPACITY = 10          # moderate capacity for all new queues

  def __init__(self):
    """Create an empty queue."""
    self._data = [None] * ArrayQueue.DEFAULT_CAPACITY
    self._size = 0
    self._front = 0

  def __len__(self):
    """Return the number of elements in the queue."""
    return self._size

  def is_empty(self):
    """Return True if the queue is empty."""
    return self._size == 0

  def first(self):
    """Return (but do not remove) the element at the front of the queue.

    Raise Empty exception if the queue is empty.
    """
    # if self.is_empty():
    #   raise Empty('Queue is empty')
    return self._data[self._front]

  def dequeue(self):
    """Remove and return the first element of the queue (i.e., FIFO).

    Raise Empty exception if the queue is empty.
    """
    # if self.is_empty():
    #   raise Empty('Queue is empty')
    answer = self._data[self._front]
    self._data[self._front] = None         # help garbage collection
    self._front = (self._front + 1) % len(self._data)
    self._size -= 1
    return answer

  def enqueue(self, e):
    """Add an element to the back of queue."""
    if self._size == len(self._data):
      self._resize(2 * len(self.data))     # double the array size
    avail = (self._front + self._size) % len(self._data)
    self._data[avail] = e
    self._size += 1

  def _resize(self, cap):                  # we assume cap >= len(self)
    """Resize to a new list of capacity >= len(self)."""
    old = self._data                       # keep track of existing list
    self._data = [None] * cap              # allocate list with new capacity
    walk = self._front
    for k in range(self._size):            # only consider existing elements
      self._data[k] = old[walk]            # intentionally shift indices
      walk = (1 + walk) % len(old)         # use old size as modulus
    self._front = 0                        # front has been realigned


if __name__ == '__main__':
  fifo1 = ArrayQueue()
  fifo2 = ArrayQueue()

  # input = [[0, 100, 20], [0, 20, 24], [0, 200, 36], [1, 150, 30]]     #0- buy, 1- sell
  # input1 = [[1, 1000, 20]]
  # for i in input:
  #   if i[0] == 0:
  #     fifo1.enqueue(i[1:])
  #   if i[0] == 1:
  #     fifo2.enqueue(i[1:])


  def Input(input):
    for i in input:
      if i[0:4] == "buy ":
        nums = [int(nums) for nums in re.findall(r'-?\d+\.?\d*', i)]      #For reference
        fifo1.enqueue(nums)
      elif i[0:4] == "sell":
        nums = [int(nums) for nums in re.findall(r'-?\d+', i)]
        fifo2.enqueue(nums)
        
    CapitalGain()


  def CapitalGain():
    global answer
    answer = 0
    
    while not(fifo2.is_empty()):
      current1 = fifo1.first()
      current2 = fifo2.first()
      
      #Error Checkings
      if fifo1.is_empty():
        print("Not enough stocks to sell")
        break

      #Capital Gain Computation
      if current2[0] - current1[0] >= 0:
        answer += current1[0] * (current2[1] - current1[1])
        current2[0] -= current1[0]
        fifo1.dequeue()
      else:
        answer += current2[0] * (current2[1] - current1[1])
        current2[0] -= current1[0]

      if current2[0] <= 0:
        fifo2.dequeue()
        
    return answer


  #Input format- ["buy/sell intx share(s) at inty each", ...], assume input format is correct
  input1 = ["buy 100 share(s) at 20 each", "buy 20 share(s) at 24 each", "buy 200 share(s) at 36 each", "sell 150 share(s) at 30 each"]
  input = ["buy 100 share(s) at 10 each", "sell 1000 share(s) at 20 each"]

  #Output
  Input(input)
  print("The total capital gain is: " + str(answer))


  #References
  #Extracting numbers from string- https://blog.finxter.com/how-to-extract-numbers-from-a-string-in-python/#:~:text=The%20easiest%20way%20to%20extract,For%20example%2C%20re.
````
